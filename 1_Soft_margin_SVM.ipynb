1. Soft margin SVM

Data set 1

```
import pandas as pd
from google.colab import files
import csv
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```
#Uploading the datas set
uploaded = files.upload()
with open("train_1.txt") as infile, open("outfile1.csv", "w") as outfile:
    csv_writer = csv.writer(outfile)
    csv_writer.writerow(['x1','x2','y'])
    for line in infile.read().splitlines():
        v=line.split()
        csv_writer.writerow([v[0],v[1],v[2]])
dftrain1 = pd.read_csv('outfile1.csv') 

uploaded = files.upload()
with open("test_1.txt") as infile, open("outfile2.csv", "w") as outfile:
    csv_writer = csv.writer(outfile)
    csv_writer.writerow(['x1','x2','y'])
    for line in infile.read().splitlines():
        v=line.split()
        csv_writer.writerow([v[0],v[1],v[2]])
dftest1 = pd.read_csv('outfile2.csv') 
```

```
xtrain1=dftrain1.iloc[:,:2]
ytrain1=dftrain1.iloc[:,2]
#SVM classifier using linear kernel
db=svm.SVC(C=1000,kernel='linear')
db.fit(xtrain1,ytrain1)
plt.scatter(dftrain1.iloc[:,0],dftrain1.iloc[:,1],c=ytrain1)

axis = plt.gca()
xlim = axis.get_xlim()
ylim = axis.get_ylim()
xaxis = np.linspace(xlim[0], xlim[1], 30)
yaxis = np.linspace(ylim[0], ylim[1], 30)
Yaxis, Xaxis = np.meshgrid(yaxis, xaxis)

XY = np.vstack([Xaxis.ravel(), Yaxis.ravel()]).T
Z = db.decision_function(XY).reshape(Xaxis.shape)
axis.contour(Xaxis, Yaxis, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
axis.scatter(db.support_vectors_[:, 0], db.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

print(db.coef_)
```

```
xtest1=dftest1.iloc[:,:2]
ytest1=dftest1.iloc[:,2]
pred=db.predict(xtest1)
n=len(ytest1)
count=0
for i in range(n):
  if ytest1[i]!=pred[i]:
    count=count+1
print("fraction of number of misclassified data points",end=': ')
print(count/n)
```

```
accuracy = accuracy_score(ytest1, pred)
print("Prediction accuracy",end=': ')
print(accuracy)
```
